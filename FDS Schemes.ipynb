Let's start with a simple sine wave scaled by $2\pi$ in the x direction with same overlayed points.

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 14
```

```python
X = np.linspace(0,2*pi,1000)
Y = np.zeros_like(X)
Y = np.sin(X)
exact = np.cos(X)

X_pts = np.linspace(0,2*pi,20)
Y_pts = np.zeros_like(X_pts)
Y_pts = np.sin(X_pts)

fig = plt.figure(figsize=(8,5))
plt.plot(X/(2*pi),Y,'k',(X_pts/(2*pi)),Y_pts,'ro')
```

## Forward Difference

### Derivatives

```python
def ForwardDiff(N):
    X = np.linspace(0,2*pi,N)
    Y = np.zeros_like(X)
    Y = np.sin(X)
    dY = np.zeros_like(Y)
    
    for i in range(0,N-1):
        dY[i] = (Y[i+1] - Y[i])/N
    
    return dY
```

```python
dY_6 = ForwardDiff(6)
X_6 = np.linspace(0,2*pi,6)
dY_16 = ForwardDiff(16)
X_16 = np.linspace(0,2*pi,16)
dY_32 = ForwardDiff(32)
X_32 = np.linspace(0,2*pi,32)
dY_64 = ForwardDiff(64)
X_64 = np.linspace(0,2*pi,64)
dY_128 = ForwardDiff(128)
X_128 = np.linspace(0,2*pi,128)
dY_256 = ForwardDiff(256)
X_256 = np.linspace(0,2*pi,256)
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(X/(2*pi),exact, 'k')
plt.plot(X_6/(2*pi),dY_6, 'bo')
plt.plot(X_16/(2*pi),dY_16, 'go')
plt.plot(X_32/(2*pi),dY_32, 'ro')
plt.plot(X_64/(2*pi),dY_64, 'co')
plt.plot(X_128/(2*pi),dY_128, 'mo')
plt.plot(X_256/(2*pi),dY_256, 'yo')
plt.show()
```

### Error

```python
fd6 = np.linalg.norm((dY_6 - np.cos(X_6)), ord=2)
fd16 = np.linalg.norm((dY_16 - np.cos(X_16)), ord=2)
fd32 = np.linalg.norm((dY_32 - np.cos(X_32)), ord=2)
fd64 = np.linalg.norm((dY_64 - np.cos(X_64)), ord=2)
fd128 = np.linalg.norm((dY_128 - np.cos(X_128)), ord=2)
fd256 = np.linalg.norm((dY_256 - np.cos(X_256)), ord=2)

N_log = [6,16,32,64,128,256]
#fd_log = 

fig = plt.figure(figsize=(8,5))
plt.loglog(6, fd6 ,color='k', ls='-', lw=2, marker='o')
plt.loglog(16,fd16, color='k', ls='-', lw=2, marker='o')
plt.loglog(32, fd32, color='k', ls='-', lw=2, marker='o')
plt.loglog(64, fd64, color='k', ls='-', lw=2, marker='o')
plt.loglog(128, fd128, color='k', ls='-', lw=2, marker='o')
plt.loglog(256, fd256, color='k', ls='-', lw=2, marker='o')

plt.loglog(fd256,256,color='c', ls='-', lw=2, marker='o')
plt.show()
```

## Backward Difference

### Derivatives

```python
def BackwardDiff(N):
    X = np.linspace(0,2*pi,N)
    Y = np.zeros_like(X)
    Y = np.sin(X)
    dY = np.zeros_like(Y)
    
    for i in range(1,N):
        dY[i] = (Y[i] - Y[i-1])/N
    
    return dY
```

```python
dY_6 = BackwardDiff(6)
X_6 = np.linspace(0,2*pi,6)
dY_16 = BackwardDiff(16)
X_16 = np.linspace(0,2*pi,16)
dY_32 = BackwardDiff(32)
X_32 = np.linspace(0,2*pi,32)
dY_64 = BackwardDiff(64)
X_64 = np.linspace(0,2*pi,64)
dY_128 = BackwardDiff(128)
X_128 = np.linspace(0,2*pi,128)
dY_256 = BackwardDiff(256)
X_256 = np.linspace(0,2*pi,256)
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(X/(2*pi),exact, 'k')
plt.plot(X_6/(2*pi),dY_6, 'bo')
plt.plot(X_16/(2*pi),dY_16, 'go')
plt.plot(X_32/(2*pi),dY_32, 'ro')
plt.plot(X_64/(2*pi),dY_64, 'co')
plt.plot(X_128/(2*pi),dY_128, 'mo')
plt.plot(X_256/(2*pi),dY_256, 'yo')
plt.show()
```
